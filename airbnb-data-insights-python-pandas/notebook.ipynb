# Data Cleaning Section

##### Price column

```python
clean_data = raw_data.copy()

# View the price column
clean_data['price'].head(10)
```

```python
# Remove dollar sign and commas from price, then convert to float datatype

clean_data['price'] = (
    raw_data['price']
    .replace(r'[\$,]', '', regex=True)
    .astype(float)
)

print(clean_data['price'])
```

##### Date column

```python
# View the date column

clean_data['last_review'].head(10)
```

```python
# Convert the date column to proper date data type

clean_data['last_review'] = pd.to_datetime(clean_data['last_review'], errors='coerce')

print(clean_data['last_review'].head(10))
```

##### Handling Missing Values

```python
# Replace missing values in 'reviews_per_month' column with 0
clean_data['reviews_per_month'] = clean_data['reviews_per_month'].fillna(0)

# Replace missing values in 'host_name' column with 'Not Specified' to indicate unavailable host information
clean_data['host_name'] = clean_data['host_name'].fillna("Not Specified")

```

##### Remove irrelevant rows (e.g., listings with zero availability or price).


```python
# Remove rows where 'price' or 'availability' contain missing values

clean_data= clean_data.dropna(subset= ['price', 'has_availability', 'availability_30','availability_60', 'availability_90', 'availability_365'])
```

# Data Enrichment Section

##### Create total cost per booking

```python
enriched_data = clean_data.copy()

# Calculate the estimated total cost per booking by multiplying nightly price by minimum nights required
enriched_data['price_per_booking'] = enriched_data['price'] * enriched_data['minimum_nights']

# Display a preview of the new calculated column
enriched_data[['price', 'minimum_nights', 'price_per_booking']].head()
```

##### Create availability categories

```python
# Categorize listings based on their yearly availability:
# 'Rare'      → properties available less than 100 days per year
# 'Part-time' → properties available between 100 and 300 days per year
# 'Full-time' → properties available between 301 and 365 days per year

enriched_data['availability_category'] = pd.cut(
    enriched_data['availability_365'], 
    bins=[-1, 99, 300, 365],
    labels=['Rare', 'Part-time', 'Full-time']
)

# Display a preview of the new availability_category column
enriched_data[['availability_365', 'availability_category']].head()

enriched_data.to_csv('enriched_uk_england_airbnb_listing_data.csv', index= False)
```

# Data Analysis Section

##### 1. What are the top 10 most expensive neighborhoods by average price ?

```python
# Calculate the top 10 most expensive neighborhoods
top_10_expensive_neighborhoods =(
     enriched_data.groupby('neighbourhood_cleansed')['price']
    .mean()
    .sort_values(ascending= False).head(10)
)
print("Top 10 most expensive neighborhoods by average price:")
print(top_10_expensive_neighborhoods)
```

##### 2. What’s the average availability and price by room type ?

```python
# Calculate the average price and availability by room type

avg_by_room_type = (
    enriched_data.groupby('room_type')[['availability_365', 'price']]
    .mean()
    .reset_index()
)

print("\nAverage price and availability by room type:")
print(avg_by_room_type)
```

##### 3. Which host has the most listings?

```python
# Host with the most listings

top_host = (
    enriched_data['host_name'].value_counts().head(1)
)

print(top_host)
```

##### 4. How does average price vary across different boroughs or districts?

```python
# Average price across boroughs/districts

avg_price_by_district = (
    enriched_data
    .groupby('neighbourhood_cleansed')['price']
    .mean()
    .reset_index()
    .sort_values(by='price', ascending=False)
)

avg_price_by_district.head()
```

##### 5. How many listings have never been reviewed ?

```python
never_reviewed = enriched_data[
    (enriched_data['number_of_reviews'].isna()) | 
    (enriched_data['number_of_reviews'] == 0)
]

# Display total count
print("Number of listings that have never been reviewed:", len(never_reviewed))
```
